## Install
---

 - First execute the cell below: with the package mananger install `pkg.installPackage(groupId:artifactId:version)` the jar and its compile dependencies. You can find that in maven repository.
 - Relaunch the kernel and the cell. You should have this output

```python
from pixiedust.packageManager import PackageManager
pkg=PackageManager()
pkg.installPackage("graphframes:graphframes:0.6.0-spark2.2-s_2.11")
pkg.installPackage("com.typesafe.scala-logging:scala-logging-api_2.11:2.1.2")
pkg.installPackage("com.typesafe.scala-logging:scala-logging-slf4j_2.11:2.1.2")
pkg.installPackage("org.scala-lang:scala-library:2.11.8")
pkg.printAllPackages()
 
sqlContext=SQLContext(sc)
```

  * If everything is gucci (good), the jars have now integrated your java.class.path

## Example

```python
#import the display module
from pixiedust.display import *
#import the Graphs example
from graphframes.examples import Graphs
#create the friends example graph
g=Graphs(sqlContext).friends()
#use the pixiedust display
display(g)
```
